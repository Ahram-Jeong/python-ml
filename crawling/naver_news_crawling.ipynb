```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 네이버 뉴스 > IT/과학 > IT 일반 뉴스 제목 크롤링
U_A = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0"}
res = req.get("https://news.naver.com/breakingnews/section/105/230", headers = U_A)
news_txt = bs(res.text, "lxml")
title = news_txt.select("a.sa_text_title")

titles = []
for i in title :
    titles.append(i.text.strip())

df_data = pd.DataFrame(titles, index = range(1, len(titles)+1))
df_data.index.name = "순서"
df_data.columns = ["제목"]
df_data
```
