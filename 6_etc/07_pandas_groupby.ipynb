### **(3) ê·¸ë£¹ í•„í„°ë§**

**`DataFrame.groupby("ì—´ ì´ë¦„").filter(ì¡°ê±´ í•¨ìˆ˜)`**

- íŠ¹ì • ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ê·¸ë£¹ì„ ì„ íƒí•œë‹¤.
- **`ì¡°ê±´ í•¨ìˆ˜`** ëŠ” ê° ê·¸ë£¹ì— ì ìš©ë˜ì–´ **`True`** ë¥¼ ë°˜í™˜í•˜ëŠ” ê·¸ë£¹ë§Œ ì„ íƒí•œë‹¤.
- ì˜ˆì‹œ ì½”ë“œ
  - ì§ì›ìˆ˜ê°€ 2ëª…ë³´ë‹¤ ë§ì€ ë¶€ì„œ í•„í„°ë§


```python
# ë¶€ì„œì˜ ì¸ì› ìˆ˜ê°€ 2ëª… ë³´ë‹¤ ë§ì€ ë¶€ì„œë¥¼ í•„í„°ë§
# 1. ê·¸ë£¹í™”
grouped = df.groupby("ë¶€ì„œ")


# ì¡°ê±´ í•¨ìˆ˜
# ë§¤ê°œ ë³€ìˆ˜ : ê·¸ë£¹í™” í•œ ë°ì´í„°ì˜ ê°œë³„ í–‰
def greater_cnt_two(group):
    return group["ì´ë¦„"].count() > 2  # ë¶€ì„œì˜ ì§ì› ìˆ˜ê°€ 2ëª…ë³´ë‹¤ ë§ì€ ë¶€ì„œ


# 2. í•„í„°ë§
grouped.filter(greater_cnt_two)
```

```python
# í‰ê·  ì—°ë´‰ì´ 6000 ì´ìƒì¸ ë¶€ì„œ í•„í„°ë§ -> ë¶€ì„œì˜ ì—°ë´‰ í‰ê· ì„ ë‚´ì—ˆì„ ë•Œ 6000 ì´ìƒ
# df.groupby("ë¶€ì„œ")["ì—°ë´‰"].mean()


def gt_sal(gr):
    return gr["ì—°ë´‰"].mean() >= 6000


df.groupby("ë¶€ì„œ").filter(gt_sal)
```

```python
# ì—°ë´‰ì˜ í‰ê· , ì¤‘ìœ„ ê°’
df["ì—°ë´‰"].agg(["median", "mean"])
```

```python
# ë‚˜ì´ì˜ ìµœì†Œ, ìµœëŒ€, í‰ê· 
df["ë‚˜ì´"].agg(["min", "max", "mean"])
```

- ê° ì—´ì— ì—¬ëŸ¬ê°œì˜ ì•„ë˜ í•¨ìˆ˜ ì ìš©
  - ì´ë¦„ : count
  - ë‚˜ì´ : min, max, mean

```python
# ì—¬ëŸ¬ ê°œì˜ ì»¬ëŸ¼ì— ì—¬ëŸ¬ ê°œì˜ ì§‘ê³„ í•¨ìˆ˜ -> ë”•ì…”ë„ˆë¦¬ì˜ í˜•íƒœë¡œ í™œìš©, key : col, value : ì§‘ê³„ í•¨ìˆ˜
# ì´ë¦„ì˜ ê°œìˆ˜
# ì—°ë´‰ì˜ í‰ê· , ì¤‘ìœ„ ê°’
# ë‚˜ì´ì˜ ìµœì†Œ, ìµœëŒ€, í‰ê· 
df.agg(
    {
        "ì´ë¦„": "count",
        "ì—°ë´‰": ["mean", "median"],
        "ë‚˜ì´": ["min", "max", "mean"],
    }
)
```

```python
# ê·¸ë£¹í™” ëœ ë°ì´í„°ì—ë„ ë‹¤ì¤‘ ì§‘ê³„ í•¨ìˆ˜ agg()
# ì»¬ëŸ¼ì´ ê³„ì¸µ êµ¬ì¡°ë¥¼ ê°–ê²Œ ë¨
df2 = df.groupby("ì§€ì—­").agg(
    {
        "ì´ë¦„": "count",
        "ì—°ë´‰": ["mean", "median"],
        "ë‚˜ì´": ["min", "max", "mean"],
    }
)
df2  # ëª¨ë“  ê³„ì¸µ êµ¬ì¡° ì¶œë ¥
df2[["ì—°ë´‰"]]  # ì—°ë´‰ì˜ ë‹¤ì¤‘ ì§‘ê³„ ì¶œë ¥
df2[[("ì—°ë´‰", "mean")]]  # ì—°ë´‰ì˜ í‰ê·  ì»¬ëŸ¼ ì¶œë ¥ -> ()ë¡œ ë¬¶ì–´ì•¼ í•¨
```
