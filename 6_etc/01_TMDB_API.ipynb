## TMDB API 실습

- 현재 상영 중인 영화 정보(results)만 추출하기
- 현재 상영 중인 영화 목록 중 첫 3개의 영화 출력하기
- 현재 상영 중인 영화 목록을 1 페이지부터 5 페이지까지 요청하고, 응답 데이터 리스트에 저장하고, 출력하기
- 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화 찾기
  - 1 페이지부터 5페이지까지의 영화 중
  - 해당 영화의 평균 평점과 영화 이름 출력
- 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화의 리뷰 데이터 요청하고, 응답 데이터 출력하기


```python
import requests as req
import json

# API KEY
with open("../data/API_KEYS.json") as f :
    api_keys = json.loads(f.read())
    
# Request Parameter
api_key = api_keys["TMDB_API_KEY"]
api_token = api_keys["TMDB_API_TOKEN"]
base_url = "https://api.themoviedb.org/3"
endpoint = "/movie/now_playing"
req_url = base_url + endpoint
params = {
    "api_key": api_key,
    "language": "ko-KR",
}

# headers
headers = {
    "accept": "application/json",  # 응답 형식
}

res = req.get(req_url, headers=headers, params=params)
print(res)
now_list = res.json()
now_list
```

```python
# 현재 상영 중인 영화 정보(results)만 추출하기
now_mv_list = now_list["results"]
# 현재 상영 중인 영화 목록 중 첫 3개의 영화 출력하기
now_mv_list[:3]
```

```python
# Request Parameter
# 현재 상영 중인 영화 목록을 1 페이지부터 5 페이지까지 요청하고, 응답 데이터 리스트에 저장하고, 출력하기
result_list = []

for i in range(1, 6):
    params = {
        "api_key": api_key,
        "language": "ko-KR",
        "page": i,
    }
    res = req.get(req_url, params=params)
    result = res.json()
    result_list.extend(result["results"])

result_list
```

```python
# 해당 영화의 평균 평점과 영화 이름 출력
print(
    f"제목 : {result_list[-1]['title']}, 평균 펑점 : {result_list[-1]['vote_average']}"
)
```

```python
# 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화 찾기
#   - 1 페이지부터 5페이지까지의 영화 중
#   - 해당 영화의 평균 평점과 영화 이름 출력
top_score = result_list[0]["vote_average"]
top_mv_info = {}

for i in result_list:
    if i["vote_average"] > top_score:
        top_score = i["vote_average"]
        top_mv_info = i

print(top_score)
print(f"제목 : {top_mv_info['title']}, 평균 펑점 : {top_mv_info['vote_average']}")
top_mv_info
```

```python
# 현재 상영 중인 영화 중 평균 평점(vote_average)이 가장 높은 영화의 리뷰 데이터 요청하고, 응답 데이터 출력하기
movie_id = top_mv_info["id"]
endpoint = f"/movie/{movie_id}/reviews?language=en-US"
req_url = base_url + endpoint
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_token}",
}
res = req.get(req_url, headers=headers)
top_mv_review = res.json()
top_mv_review  # 리뷰 데이터 없음
```

```python
# Top Rated API의 첫 번째 영화 ID로 영화 리뷰 데이터 요청하고, 응답 데이터 출력하기
endpoint = f"/movie/top_rated?language=ko-KR"
req_url = base_url + endpoint
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_token}",
}
res = req.get(req_url, headers=headers)
top_rated_mv = res.json()
top_rated_mv
```

```python
# Top Rated API의 첫 번째 영화 리뷰 데이터
movie_id = top_rated_mv["results"][0]["id"]
endpoint = f"/movie/{movie_id}/reviews"
req_url = base_url + endpoint
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {api_token}",
}
res = req.get(req_url, headers=headers)
top_rated_review = res.json()
top_rated_review
```
