```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 네이버 증권 > 시장지표 > 환율 리스트 크롤링 (통화명, 매매기준율)
U_A = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0"}
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver", headers = U_A)
html = bs(res.text, "lxml")
curr = html.select("td.tit > a")
sale = html.select("td.sale")
len(curr) # 58
len(sale) # 58

curr_lst = []
sale_lst = []
for i in range(len(curr)) :
    curr_lst.append(curr[i].text.strip())
    sale_lst.append(sale[i].text.strip() + " 원")

result = {"통화명" : curr_lst, "매매기준율" : sale_lst}

df = pd.DataFrame(result, index = range(1, len(curr)+1))
df
```
